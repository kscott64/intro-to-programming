```
import random
n=random.randint(1,10)
g=0
a=input("What is your name;")
print("Hello",a,"Guess a number between 1 and 10")
while g<3:
  u=int(input("Enter your guess;"))
  g=g+1
  if u==n:
    print(a,"You won!")
    break
  elif u<n:
    print("Too low")
  else:
    print("Too high")
if u!=n:
  print("Game Over. The number was",n)
```
